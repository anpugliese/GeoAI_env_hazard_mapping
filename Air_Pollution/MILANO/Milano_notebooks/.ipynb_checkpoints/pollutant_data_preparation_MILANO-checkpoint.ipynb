```python
import pandas as pd
import geopandas as gpd
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt

import interpolation_module as interp
milano_shapefile_path = '../../data/MilanoComune/comuneMilano.shp'
```

```python
no2_timeseries = pd.read_csv("../../data/milano_pollutants_data/no2_hourly.csv", index_col=0)
so2_timeseries = pd.read_csv("../../data/milano_pollutants_data/so2_hourly.csv", index_col=0)
pm10_timeseries = pd.read_csv("../../data/milano_pollutants_data/pm10_hourly.csv", index_col=0)
o3_timeseries = pd.read_csv("../../data/milano_pollutants_data/o3_hourly.csv", index_col=0)
pm25_timeseries = pd.read_csv("../../data/milano_pollutants_data/pm25_hourly.csv", index_col=0)
```

```python
#Function to get the records in which a set threshold is exceeded hourly
def get_hourly_exceeded(timeseries, threshold, value_column='value'):
    timeseries_hourly_means = timeseries.copy()
    timeseries_hourly_exceeded = timeseries_hourly_means[(timeseries_hourly_means[value_column] >= threshold)]
    
    #get the exceeded by year counts
    yearly_counts = timeseries_hourly_exceeded.copy()
    yearly_counts['date'] = pd.to_datetime(yearly_counts['date'], format = '%Y-%m-%d')
    yearly_counts = yearly_counts.groupby([yearly_counts['date'].dt.year, 'lat', 'lng', 'sensor_id']).count()
    
    return timeseries_hourly_exceeded, yearly_counts
```

```python
#Function to get the records in which a set threshold is exceeded daily
def get_daily_exceeded(timeseries, threshold, value_column='value'):
    timeseries_daily_means = timeseries.copy()
    timeseries_daily_means['date'] = pd.to_datetime(timeseries_daily_means['date'], format = '%Y-%m-%d')
    
    #group by day
    timeseries_daily_means = timeseries_daily_means.groupby([timeseries_daily_means['date'].dt.date, 'lat', 'lng', 'sensor_id']).mean()    
    timeseries_daily_exceeded = timeseries_daily_means[(timeseries_daily_means[value_column] >= threshold)]
    
    #get the exceeded by year counts
    yearly_counts = timeseries_daily_exceeded.copy()
    yearly_counts = yearly_counts.reset_index(level='date', drop=False)
    yearly_counts['date'] = pd.to_datetime(yearly_counts['date'], format = '%Y-%m-%d')
    yearly_counts = yearly_counts.groupby([yearly_counts['date'].dt.year, 'lat', 'lng', 'sensor_id']).count()
    
    return timeseries_daily_exceeded, yearly_counts
```

```python
def show_plot(df, x_col, y_col, name="Unnamed Plot", additional_traces=[]):
    if type(df) != list:
        df = [df]
        x_col = [x_col]
        y_col = [y_col]
        name = [name]
    
    fig = go.Figure()
    for i, single_df in enumerate(df):
        x = single_df[x_col[i]]
        y = single_df[y_col[i]]
        fig_name = name[i]
        fig.add_trace(go.Scatter(x=x, y=y, mode='lines+markers',name=fig_name))
    
    if len(additional_traces) > 0:
        for trace in additional_traces:
            fig.add_trace(trace)
    fig.show()
```

```python
"""
Value thresholds for no2
"""
no2_timeseries
no2_timeseries = no2_timeseries.groupby(['date', 'lat', 'lng']).mean()
no2_timeseries = no2_timeseries.reset_index()
```

```python
"""
Pollutant   hourly   hours/y   daily    days/y
NO2         200      18        40       NS
"""
no2_hourly_threshold = 200 #ug/m3
no2_daily_threshold = 40 #ug/m3
no2_hourly_exceeded, no2_hourly_exceeded_counts = get_hourly_exceeded(no2_timeseries, no2_hourly_threshold)
no2_daily_exceeded, no2_daily_exceeded_counts = get_daily_exceeded(no2_timeseries, no2_daily_threshold)
```

```python
"""
Value thresholds for so2
"""
so2_timeseries
so2_timeseries = so2_timeseries.groupby(['date', 'lat', 'lng']).mean()
so2_timeseries = so2_timeseries.reset_index()
```

```python
"""
Pollutant   hourly   hours/y   daily    days/y
SO2         350      24        125       3
"""
so2_hourly_threshold = 350 #ug/m3
so2_daily_threshold = 125 #ug/m3
so2_hourly_exceeded, so2_hourly_exceeded_counts = get_hourly_exceeded(so2_timeseries, so2_hourly_threshold)
so2_daily_exceeded, so2_daily_exceeded_counts = get_daily_exceeded(so2_timeseries, so2_daily_threshold)
```

```python
"""
Value thresholds for pm10
"""
#remove duplicates by doing the mean in observations on the same datetime and location
pm10 = pm10_timeseries.copy()
pm10
pm10 = pm10.groupby(['date', 'sensor_id', 'lat', 'lng']).mean()
pm10 = pm10.reset_index()
```

```python
pm10
```

```python
"""
Pollutant   daily     exceedances per year
PM10        50        35
"""
pm10_daily_threshold = 50 #ug/m3
pm10_daily_exceeded, pm10_daily_exceeded_counts = get_daily_exceeded(pm10, pm10_daily_threshold)
```

```python
"""----------------------------------------------------------------------------------------------"""
```

```python
# Data analysis for PM10 daily data
```

```python
#get the station ids
stations = pm10.date.unique()
stations = list(stations)
print(len(stations))
```

```python
for sensor_id in stations:
    station_data = pm10.loc[pm10['sensor_id'] == sensor_id]
    additional_traces = [
        go.Scatter(
            x=[station_data.iloc[0].date, station_data.iloc[-1].date], 
            y=[pm10_daily_threshold]*2, 
            mode='lines+markers',
            name="Threshold"
        )
    ]
    show_plot(station_data, "date", "value", f"PM10 station {sensor_id}", additional_traces=additional_traces)
    
```

```python
#Calculate exceedance rate
pm10_data_per_year = pm10.copy()
pm10_data_per_year['date'] = pd.to_datetime(pm10_data_per_year['date'], format = '%Y-%m-%d')
pm10_data_per_year = pm10_data_per_year.groupby([pm10_data_per_year['date'].dt.year, 'lat', 'lng', 'sensor_id']).count()
pm10_data_per_year = pm10_data_per_year.drop(['date'], axis=1)
pm10_data_per_year = pm10_data_per_year.rename(columns={"value": "days_with_data"})
pm10_data_per_year["exceeded"] = pm10_daily_exceeded_counts.value
pm10_data_per_year["exceeded_rate"] = (pm10_data_per_year["exceeded"] / pm10_data_per_year["days_with_data"] * 100)

pm10_data_per_year
```

```python
plot_stations = []
plot_stations_x = []
plot_stations_y = []
plot_stations_name = []
for sensor_id in stations:
    station_data = pm10_data_per_year.reset_index()
    station_data = station_data.loc[station_data['sensor_id'] == sensor_id]
    plot_stations.append(station_data)
    plot_stations_x.append("date")
    plot_stations_y.append("exceeded_rate")
    plot_stations_name.append(f"PM10 station {sensor_id}")
    
show_plot(plot_stations, plot_stations_x, plot_stations_y, name=plot_stations_name)
```

```python
pm10_data_plot = pm10.copy()
epsg_wgs = 4326

#Plot of Milano to see where the stations are and their labels
bnd = gpd.read_file(milano_shapefile_path)
fig, ax = plt.subplots(1,1, figsize=(15,15))
ax.ticklabel_format(useOffset=False)
bnd.plot(ax=ax, facecolor='w', edgecolor='k')

#coordinates of sensors
sensor_coords = pm10_data_plot.groupby(['lat','lng', 'sensor_id'], as_index=True).mean().reset_index().drop(columns=['value'])
coordinates_ = gpd.GeoDataFrame(sensor_coords)
coordinates_['geometry'] = gpd.points_from_xy(
    sensor_coords['lng'], 
    sensor_coords['lat'],
    crs = epsg_wgs
)
coordinates_.plot(ax=ax)

for i in coordinates_.index:
    plt.annotate(coordinates_["sensor_id"][i], (coordinates_["lng"][i], coordinates_["lat"][i]))

plt.show()
```
