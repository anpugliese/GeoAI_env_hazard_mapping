```python
import xarray as xr
import numpy as np
import pandas as pd
import geopandas as gpd
import datacube
from sklearn.preprocessing import OneHotEncoder
```

```python
# Get the list of locations of the different stations for air quality pollutants and meteo
meteo_complete = pd.read_csv('../../data/milano_meteo_data/meteo_data_merged_daily_interpolated.csv',index_col=0)
pm10_data = pd.read_csv('../../data/milano_pollutants_data/pm10_daily.csv',index_col=0)
#no2_data = pd.read_csv("../../data/milano_pollutants_data/no2_hourly.csv", index_col=0)
#so2_data = pd.read_csv("../../data/milano_pollutants_data/so2_hourly.csv", index_col=0)
#o3_data = pd.read_csv("../../data/milano_pollutants_data/o3_hourly.csv", index_col=0)
#pm25_data = pd.read_csv("../../data/milano_pollutants_data/pm25_hourly.csv", index_col=0)
```

```python
meteo_stations_location = meteo_complete[['lat', 'lng']].drop_duplicates()
additional_locations = [
    pm10_data[['lat', 'lng']].drop_duplicates(),
    #no2_data[['lat', 'lng']].drop_duplicates()
    #so2_data[['lat', 'lng']].drop_duplicates()
    #o3_data[['lat', 'lng']].drop_duplicates()
    #pm25_data[['lat', 'lng']].drop_duplicates()
]
```

```python
print("meteo stations: " + str(len(meteo_stations_location)))
for location in additional_locations:
    print("additional locations: " + str(len(location)))
#print("pm10 stations: " + str(len(pm10_data_locations)))
```

```python
all_stations = meteo_stations_location.copy()
for additional in additional_locations:
    all_stations = pd.concat([all_stations, additional])
    
all_stations = all_stations.round({'lat': 6, 'lng': 6})

all_stations = all_stations.drop_duplicates()
len(all_stations)
```

```python
#Convert from EPSG:4326 (lat,lng) to EPSG:32632 (Italy UTM) for the geodataframe
epsg_wgs = 4326
coordinates_utm = gpd.GeoDataFrame(all_stations)
coordinates_utm['geometry'] = gpd.points_from_xy(all_stations['lng'], all_stations['lat'], crs = epsg_wgs)
coordinates_utm = coordinates_utm.to_crs({'init': 'epsg:32632'})
```

```python
# set the columns with the X and Y coordinates in UTM
coordinates_utm['utm_x'] = coordinates_utm['geometry'].x
coordinates_utm['utm_y'] = coordinates_utm['geometry'].y
#coordinates_utm = coordinates_utm.drop(columns=['utm'])
coordinates_utm
```

```python
dc = datacube.Datacube(app = "my_app", config = "/home/user/ODC_harmonia/datacube.conf")
products = dc.list_products()
products
```

```python
products.index
```

```python
# get unique lithology values
geo_datasets = dc.find_datasets(product='geologia')
geo = dc.load(datasets=geo_datasets)
geo_data = geo.codice.squeeze().values
geo_data = geo_data.flatten()
# geo_data = geo_data.reshape(-1,1)
geo_cat = np.unique(geo_data)
geo_cat = geo_cat.reshape(-1,1)[:-1]
del geo_data

# get unique land cover values
lc_datasets = dc.find_datasets(product='dusaf')
lc = dc.load(datasets=lc_datasets)
lc_data = lc.codice.squeeze().values
lc_data = lc_data.flatten()
# geo_data = geo_data.reshape(-1,1)
lc_cat = np.unique(lc_data)
lc_cat = lc_cat.reshape(-1,1)[:-1]
del lc_data
```

```python
"""
0 - ghiaie, sabbie e limi
1 - ghiaie, sabbie
2 - ghiaie, sabbie e argille ferrettizzate
3 - argille, calcari, conglomerati
4 - ghiaie, limi e argille fortemente ferrettizzati
5 - conglomerati, sabbie, argille
6 - marne
"""

# map unique values to specified lithology classes

geo_cat_dict = {
    101 : 0,
    201 : 0,
    301 : 0,
    205 : 1,
    206 : 2,
    76  : 3,
    207 : 4,
    8   : 5,
    81  : 6
}

mapped_geo_cat = np.vectorize(geo_cat_dict.get)(geo_cat.flatten())
mapped_geo_cat = np.unique(mapped_geo_cat).reshape(-1,1)
mapped_geo_cat
```

```python
# map land cover values to reduced land cover classes
def map_land_cat(value):
    value_str = str(int(value))
    value_cat = value_str[:2]
    if value_cat == '13':
        value_cat = '12'
    return int(value_cat)
```

```python
mapped_lc_cat = np.vectorize(map_land_cat)(lc_cat.flatten())
mapped_lc_cat = np.unique(mapped_lc_cat).reshape(-1,1)
mapped_lc_cat
```

```python
geo_enc = OneHotEncoder()
geo_enc.fit(mapped_geo_cat)
geo_enc.categories_
```

```python
lc_enc = OneHotEncoder()
lc_enc.fit(mapped_lc_cat)
lc_enc.categories_
```

```python
# create dataframe with all selected conditioning factors values corresponding to the selected coordinates
cf_list = ['dtm_milan','aspect','dusaf15','geologia','hillshade','ndvi_2019','plan_curvature','profile_curvature',
           'water_distance','slope','spi','tri','twi','densita_popolazione','main_road_distance','building_height']

topo_dataset = None
for cf in cf_list:
    datasets = dc.find_datasets(product=cf)
    cf_data = dc.load(datasets=datasets)
    cf_sel = cf_data.squeeze().sel(y=xr.DataArray(coordinates_utm.geometry.y.values, dims = ['index']), x = xr.DataArray(coordinates_utm.geometry.x.values, dims=['index']), method='nearest')
    cf_var_name = list(cf_data.data_vars.keys())[0]
    del cf_data
    cf_df = cf_sel.to_dataframe()
    cf_df.rename(columns={cf_var_name:cf},inplace=True)
    cf_df.drop(['time','spatial_ref'],axis=1,inplace=True)
    #cf_df
    if topo_dataset is None:
        topo_dataset = cf_df.copy()
    else:
        topo_dataset = pd.concat([topo_dataset, cf_df[cf]], axis=1)
    print(cf + ' done ' + 'length: ' + str(len(cf_df)))
    
topo_dataset
```

```python
topo_dataset_copy = topo_dataset.copy()
#topo_dataset = topo_dataset_copy.copy()

coordinates_utm_copy = coordinates_utm.copy()
#coordinates_utm = coordinates_utm_copy.copy()
```

```python
# map land cover and lithology values to the new specified classes
topo_dataset['geologia'] = np.vectorize(geo_cat_dict.get)(topo_dataset['geologia'])
topo_dataset['dusaf15'] = np.vectorize(map_land_cat)(topo_dataset['dusaf15'])
```

```python
# one hot encoding of lithology
encoded = geo_enc.transform(topo_dataset['geologia'].to_numpy().reshape(-1,1))
encoded_lith_df = pd.DataFrame(encoded.toarray(), columns=geo_enc.get_feature_names_out(['geo']))
encoded_lith_df
```

```python
# one hot encoding of land cover
encoded_lc = lc_enc.transform(topo_dataset['dusaf15'].to_numpy().reshape(-1,1))
encoded_lc_df = pd.DataFrame(encoded_lc.toarray(), columns=lc_enc.get_feature_names_out(['lc']))
encoded_lc_df
```

```python
# add the one hot encoded columns
final_topo_dataset = topo_dataset.merge(encoded_lith_df, left_index=True, right_index=True)
final_topo_dataset = final_topo_dataset.merge(encoded_lc_df, left_index=True, right_index=True)
final_topo_dataset
```

```python
coordinates_utm = coordinates_utm.reset_index()
final_topo_dataset = pd.concat([coordinates_utm, final_topo_dataset], axis=1)
final_topo_dataset = final_topo_dataset.drop(columns=['index', 'dusaf15', 'geologia', 'x', 'y'])
final_topo_dataset
```
