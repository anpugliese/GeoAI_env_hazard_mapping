```python
'''
1-11-> Residential urban areas 
2-121,13->Industrial and abbandoned urban areas
3-122,123,124 Transportation infrastructure (streets, highways, airports, and ports)
4-14->Urban green areas
5-2->Agricultural areas
6-3->Forest
7-4/5->Hydro and humid bodies
'''
#Convert from copernicus code 2018 to an internal code
URBAN = 1
INDUSTRIAL = 2
TRANSPORTATION = 3
URBAN_VEGETATION = 4
RURAL = 5
FOREST = 6
WATER = 7
LC_NO_DATA = 9999

NO_DATA = -9999
    
# Function to check if the file is a tiff and must be read.
def check_wrong_files(f):
    if f == 'clip': return True #avoid entering the "clip" folder
    if 'ipynb' in f: return True #avoid entering the "ipynb_checkpoint" file
    if 'tar' in f: return True #avoid entering "tar" files
    if 'aux' in f: return True #avoid entering "aux" files
    return False
```

```python
def match_landsat_to_landcover(landsat):
    year = int(landsat[17:21])
    if year in [2015,2016]:
        return str(2015)
    elif year in [2017,2018,2019]:
        return str(2018)
    elif year in [2020,2021,2022]:
        return str(2021)
```

```python
def extract_bits(array, bit_positions):
    """Extracts the given bits from the QA_PIXEL band."""
    mask = np.zeros_like(array, dtype=bool)
    for bit in bit_positions:
        mask |= (array & (1 << bit)) > 0  # Check if bit is set to 1
    return mask
```

```python
bands = [2, 3, 4, 5, 6, 7, 10]

def read_band(filepath, scale_factor=0.0000275, offset=-0.2, nodata_value=-9999, dtype=np.float32):
    """Reads a raster band and applies scale and offset if needed."""
    with rasterio.open(filepath, driver="GTiff") as band:
        data = band.read(1).astype(dtype)
        data = np.where(data != 0, data * scale_factor + offset, nodata_value)
        profile = band.profile.copy()
        profile.update(dtype=dtype, nodata=nodata_value, count=1)
    return data, profile

def save_raster(filepath, data, profile, data_type, no_data,band_raster=True):
    """Saves a processed raster to file."""
    profile['dtype'] = data_type
    profile['nodata'] = no_data
    os.makedirs(os.path.dirname(filepath), exist_ok=True)
    with rasterio.open(filepath, 'w', **profile) as dest:
        if band_raster:
            dest.write(data)
        else:
            dest.write(data, 1)

for f in landsat_raster_file_list:
    if check_wrong_files(f): 
        continue    
    print(f)

    year = match_landsat_to_landcover(f)
    landcover_path = f'{landcover_base_path}/DUSAF_{year}_MCM_mapped.tif'

    # Read landcover extent
    with rasterio.open(landcover_path, driver="GTiff") as landcover_raster:
        ref_geom = [box(*landcover_raster.bounds).__geo_interface__]
        landcover_raster_profile = landcover_raster.profile.copy()

    # Read cloud/shadow mask from QA band
    qa_path = f"{landsat_raster_folder}/{f}/{f}_QA_PIXEL.TIF"
    with rasterio.open(qa_path, driver="GTiff") as landsat_band_raster:
        target_masked, _ = mask(landsat_band_raster, ref_geom, crop=True)
        cloud_shadow_mask = extract_bits(target_masked[0], [3, 5])
    
    # Process bands
    band_data = {}
    for band_n in bands:
        band_prefix = 'R' if band_n != 10 else 'T'
        band_path = f"{landsat_raster_folder}/{f}/{f}_S{band_prefix}_B{band_n}.TIF"

        with rasterio.open(band_path, driver="GTiff") as landsat_band_raster:
            target_masked, _ = mask(landsat_band_raster, ref_geom, crop=True)
            target_masked = target_masked.astype(np.uint16)
            target_masked[:, cloud_shadow_mask] = 0  # Apply cloud/shadow mask
            band_data[band_n] = target_masked

        # Save clipped band
        save_raster(f"{landsat_raster_folder}/clip/{f}/{f}_S{band_prefix}_B{band_n}.TIF", 
                    target_masked, landcover_raster_profile, np.uint16, 0)

    print("Finished clipping Landsat images.")

    # **Compute LST**
    lst = np.where(band_data[10] != 0, band_data[10] * 0.00341802 + 149, NO_DATA).astype(np.float32)
    save_raster(f"{landsat_raster_folder}/clip/{f}/{f}_LST.TIF", lst, landcover_raster_profile, np.float32, NO_DATA)
    print("Finished LST")

    # **Compute NDVI** (Normalized Difference Vegetation Index)
    s4, _ = read_band(f"{landsat_raster_folder}/clip/{f}/{f}_SR_B4.TIF")
    s5, _ = read_band(f"{landsat_raster_folder}/clip/{f}/{f}_SR_B5.TIF")
    ndvi = np.where(((s5 + s4) != 0) & (s5 != -9999) & (s4 != -9999), (s5 - s4) / (s5 + s4), -9999)
    save_raster(f"{landsat_raster_folder}/clip/{f}/{f}_NDVI.TIF", ndvi, landcover_raster_profile, np.float32, NO_DATA, False)
    print("Finished NDVI")

    # **Compute NDBI** (Normalized Difference Built-up Index)
    s6, _ = read_band(f"{landsat_raster_folder}/clip/{f}/{f}_SR_B6.TIF")
    ndbi = np.where(((s6 + s5) != 0) & (s5 != -9999) & (s6 != -9999), (s6 - s5) / (s6 + s5), -9999)
    save_raster(f"{landsat_raster_folder}/clip/{f}/{f}_NDBI.TIF", ndbi, landcover_raster_profile, np.float32, NO_DATA, False)
    print("Finished NDBI")

    # **Compute Albedo**
    s2, _ = read_band(f"{landsat_raster_folder}/clip/{f}/{f}_SR_B2.TIF")
    s7, _ = read_band(f"{landsat_raster_folder}/clip/{f}/{f}_SR_B7.TIF")
    albedo = np.where((s2 != -9999) & (s4 != -9999) & (s5 != -9999) & (s6 != -9999) & (s7 != -9999),
                  (0.356*s2 + 0.130*s4 + 0.373*s5 + 0.085*s6 + 0.072*s7 - 0.018),
                  -9999)
    save_raster(f"{landsat_raster_folder}/clip/{f}/{f}_albedo.TIF", albedo, landcover_raster_profile, np.float32, NO_DATA, False)
    print("Finished Albedo")
```

```python
#build the training samples and the predict dataframe
print(f'Processing for {city}')    

with rasterio.open(landcover_path, driver="GTiff") as lancover_raster:
    landcover_array = lancover_raster.read(1)
    print(lancover_raster.profile)
    rows, cols = landcover_array.shape
    x_positions = np.arange(0, cols)
    y_positions = np.arange(0, rows)
    x, y = np.meshgrid(x_positions, y_positions)
    x_flat = x.flatten()
    y_flat = y.flatten()
    values_flat = landcover_array.flatten()

    # Create a DataFrame for the Landcover 
    landcover_df = pd.DataFrame({'x': x_flat, 'y': y_flat, 'landcover': values_flat})
    landcover_df['landcover'] = landcover_df['landcover']#.apply(map_urban_atlas_class).astype('int16')

# Process each satellite image
sat_images_path = f"{data_folder}/clip"
file_list = os.listdir(f"{sat_images_path}")
samples = pd.DataFrame()
predict_lst = np.zeros_like(landcover_array)
predict_ndvi = np.zeros_like(landcover_array)
predict_n = 0

sample_n = [
    int(total_samples_per_raster / 4), # urban, uhi 1
    int(total_samples_per_raster / 4), # urban, uhi 0
    int(total_samples_per_raster / 4), # rural/vegetation/bareland, uhi 1
    int(total_samples_per_raster / 4), # rural/vegetation/bareland, uhi 0
]

for f in file_list:
    if check_wrong_files(f): continue

    print(f'Processing {f}')
    file_date_string = f.split('_')[3] #example: LC08_L2SP_194028_20160825_20200906_02_T1_LST
    year = int(file_date_string[:4])

    #columns in the end: x,y,landcover,ndvi,raster
    train_df = landcover_df.copy()

    #add the uhi column
    with rasterio.open(f"{sat_images_path}/{f}/{f}_uhi.tif", driver="GTiff") as uhi_raster:
        uhi_array = uhi_raster.read(1) #UHI band
        uhi_flat = uhi_array.flatten()
        train_df['uhi'] = pd.Series(uhi_flat).astype('int16')

    #add the ndvi column
    with rasterio.open(f"{sat_images_path}/{f}/{f}_NDVI.TIF", driver="GTiff") as ndvi_raster:
        ndvi_array = ndvi_raster.read(1) #UHI band
        ndvi_flat = ndvi_array.flatten()
        train_df['ndvi'] = pd.Series(ndvi_flat).astype('float32')

    #add the LST column
    with rasterio.open(f"{sat_images_path}/{f}/{f}_LST.TIF", driver="GTiff") as lst_raster:
        lst_array = lst_raster.read(1) #UHI band
        lst_flat = lst_array.flatten()
        train_df['lst'] = pd.Series(lst_flat).astype('float32')

    if year >= 2020:
        predict_n += 1
        predict_lst = np.where(landcover_array != -9999, (predict_lst + lst_array), -9999)
        predict_ndvi = np.where(landcover_array != -9999, (predict_ndvi + ndvi_array), -9999)


    train_df['raster'] = int(file_date_string)

    #remove nodata (-9999) from the dataframe
    train_df = train_df.loc[
        (train_df['landcover'] != -9999)
    ]

    # sample the raster
    raster_samples = pd.DataFrame()

   #urban, uhi = 0
    condition = (train_df['landcover'] == URBAN) & (train_df['uhi'] == 0) 
    sampling = train_df.loc[
        condition
    ].sample(n=sample_n[0])
    samples = pd.concat([samples, sampling])

    #urban, uhi = 1
    condition = (train_df['landcover'] == URBAN) & (train_df['uhi'] == 1) 
    sampling = train_df.loc[
        condition
    ].sample(n=sample_n[1])
    samples = pd.concat([samples, sampling])

    #rural/forest/bareland, uhi = 0
    condition = (
        ((train_df['landcover'] == VEGETATION) | 
         (train_df['landcover'] == RURAL) | 
         (train_df['landcover'] == BARELAND)) & 
         (train_df['uhi'] == 0)
    ) 
    sampling = train_df.loc[
        condition
    ].sample(n=sample_n[2])
    samples = pd.concat([samples, sampling])

    #rural/forest/bareland, uhi = 1
    condition = (
        ((train_df['landcover'] == VEGETATION) | 
         (train_df['landcover'] == RURAL) | 
         (train_df['landcover'] == BARELAND))& 
         (train_df['uhi'] == 1)
    ) 
    sampling = train_df.loc[
        condition
    ].sample(n=sample_n[3])
    samples = pd.concat([samples, sampling])


samples_base_path = f'training_samples'    
# create the "training_samples" folder if it does not exist
os.makedirs(f"{samples_base_path}", exist_ok=True)

samples_path = f'{samples_base_path}/{city}_samples.csv'
print(f'Saving samples in {samples_path}')

samples_to_save = samples.copy()
samples_to_save = samples_to_save.reset_index(drop=True)
samples_to_save.to_csv(samples_path)
samples

# save the lst and ndvi predict
if predict_n == 0: predict_n = 0.0000001
predict_lst = np.where(landcover_array != -9999, (predict_lst / predict_n), -9999)
predict_ndvi = np.where(landcover_array != -9999, (predict_ndvi / predict_n), -9999)

predict_df = pd.DataFrame({'x': x_flat, 'y': y_flat})
predict_df['landcover'] = pd.Series(landcover_array.flatten()).astype('int32')
predict_df['x'] = predict_df['x'].astype('uint32')
predict_df['y'] = predict_df['y'].astype('uint32')
predict_df['lst'] = predict_df['lst'].astype('float32')
predict_df['ndvi'] = predict_df['ndvi'].astype('float32')

predict_df = predict_df.loc[
    (predict_df['landcover'] != -9999) & (predict_df['lst'] != 0) & (predict_df['ndvi'] != 0)
]

predict_path = f'{samples_base_path}/{city}_predict.csv'
print(f'Saving predict in {predict_path}')
predict_df = predict_df.reset_index(drop=True)
predict_df.to_csv(predict_path)
    
```
